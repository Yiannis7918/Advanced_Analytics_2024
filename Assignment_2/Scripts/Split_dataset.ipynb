<h1> Perform train/validation/test split </h1>

As the professor suggests to perform a split of the dataset at the level of the games, I will do it here "manually"

I don't understand why there would be a data leakage if we do not perform the split at the level of the game

```python
import numpy as np
import random
import shutil
```

```python
#import the metadata.csv as list
import csv

def load_csv_as_list(file_path):
    data = []
    with open(file_path, 'r') as file:
        csv_reader = csv.reader(file)
        for row in csv_reader:
            data.append(row)
    return data

# Example usage:
file_path = '..//datasets//metadata.csv'
data_list = load_csv_as_list(file_path)

data_list.pop(0) #remove first element that contains the labels of the columns

print(data_list)
```

```python
#analyze the number of different games

games = []

for row in data_list:
    if row[2] not in games:
        games.append(row[2])

print(games)

print(f"Number of images: {len(data_list)}")
print(f"Number of different games: {len(games)}")
```

```python
#Visualize the distribution of the number of images per game

nbr_per_game = np.zeros(len(games))
names_per_game = [[] for i in range(len(games))]

for row in data_list:
    index_game_name = games.index(row[2])
    nbr_per_game[index_game_name] += 1 #increase counter to the game type
    names_per_game[index_game_name].append(row[0]) #append game name

print(nbr_per_game)
print(names_per_game)
```

I will try to put approximately 80% of the images into the training dataset, 10% into the test and 10% into the validation. </br>
And respect this proportion for every game.

```python
#assign image paths to train-validation-test

source_path = "E://Dataset//images//square//" #contains all the images
destination_path = "E://Dataset//images-processed//" #contains the train-validation-test folders

#assign every image of every game to the train, validation or test folder respecting the 80-10-10 split

for i in range(len(games)):
    name_game = games[i] #string, name of the game
    number_game = nbr_per_game[i] #number of images for this game

    #unit of images for validation and test set
    unit = np.ceil(number_game/10) #round up with ceil function, sure to have 1 validation and test image per game
    #minimum number of images for a game is 4, so divided by 10 gives 2-1-1

    #unit = 10%: 10% to validation (1 unit), 10% to test (1 unit) and the rest to the train

    images_game = names_per_game[i].copy()

    #shuffle the names of the images for this game
    random.shuffle(images_game) #in place random shuffling

    counter = 0
    for image in images_game:

        #add in the validation folder
        if counter < unit:
            shutil.copy2(source_path + image, destination_path + "validation//" + image)
        #add in the test folder
        elif counter < 2*unit:
            shutil.copy2(source_path + image, destination_path + "test//" + image)
        #add in the train folder
        else:
            shutil.copy2(source_path + image, destination_path + "train//" + image)
        counter += 1
    
    
```
