<h1> Statistics about the dataset </h1>

```python
import json
import numpy as np
import matplotlib.pyplot as plt
from statistics import mode
```

```python
# Path to your JSON file
json_file_path = "../datasets/dataset.json"

# Load JSON data into a dictionary
with open(json_file_path, 'r') as json_file:
    labels_dict = json.load(json_file)
```

```python
print(f"number of games: {len(labels_dict)}")
```

```python
number_screenshots_per_game = []

for element in labels_dict:
    #print(len(element["screenshots"]))
    number_screenshots_per_game.append(len(element["screenshots"]))
```

```python
print(number_screenshots_per_game)

mode_value = mode(number_screenshots_per_game)
print(f"most frequent value: {mode_value}")
print(f"number of times the most frequent value is encountered: {number_screenshots_per_game.count(mode_value)}")
```

```python
print(f"number of games with less than 5 screenshots : {np.sum(np.array(number_screenshots_per_game)<5)}")
```

```python
print(f"total number of games: {np.sum(number_screenshots_per_game)}")
print(f"Average number of screenshots per game: {np.round(np.mean(number_screenshots_per_game),2)}")
```

```python
number_screenshots_per_game_adjusted = [20 if x >=20 else x for x in number_screenshots_per_game]
```

```python
# Plot the histogram
plt.hist(number_screenshots_per_game, bins=30, color='skyblue', edgecolor='black')  # Adjust the number of bins as needed
plt.xlabel('Number of screenshots')
plt.ylabel('Frequency')
plt.title('Distribution of the number of screenshots per game')
plt.grid(True)
plt.savefig("Histogram_games.jpg")
plt.show()
```

```python
# Plot the histogram
plt.hist(number_screenshots_per_game_adjusted, bins=19, color='skyblue', edgecolor='black')  # Adjust the number of bins as needed
plt.xlabel('Number of screenshots')
plt.ylabel('Frequency')
plt.title('Distribution of the number of screenshots per game adjusted')
plt.grid(True)
# Set x-axis ticks from 1 to 20
plt.xticks(np.arange(1, 21, step=1))
plt.savefig("Histogram_games_adjusted.jpg")
plt.show()
```
