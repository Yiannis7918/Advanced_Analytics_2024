Show multiple images from the dataset

```python
#import the metadata.csv as list
import csv

def load_csv_as_list(file_path):
    data = []
    with open(file_path, 'r') as file:
        csv_reader = csv.reader(file)
        for row in csv_reader:
            data.append(row)
    return data

# Example usage:
file_path = '..//datasets//metadata.csv'
data_list = load_csv_as_list(file_path)
print(data_list)
```

```python
#plot a few images from the dataset
#select first 9 images of the list

image_paths = data_list[1:10]

print(image_paths)
```

```python
#plot the 9 images
plt.figure(figsize=(10, 6)) #size of the window

for i, list_path in enumerate(image_paths):
    plt.subplot(3, 3, i + 1)  # Adjust subplot to 3x3 grid
    img = image.load_img("G:\\Dataset\\images\\"+list_path[0], target_size=(1080,1920))
    plt.imshow(img)
    plt.axis('off')

plt.tight_layout()  # Adjust spacing between subplots
plt.savefig("6_Knuckle_sandwich_3_kingdom-heroes_bottom")
plt.show()
```

```python
print(data_list[0])
len(data_list)

data_array = np.array(data_list)
```

```python
# Find the index of the row with the maximum value in column 'x'
max_row_index = np.argmax(data_array[1:len(data_array),1])

print(max_row_index)

# Extract the value from column 'y' in the row with the maximum value in column 'x'
y_value = data_array[max_row_index, 0]

print(y_value)
print(data_array[max_row_index,1])
```

```python
price_list = data_array[1:,1]
name_list = data_array[1:,0]
game_list = data_array[1:,2]
```

```python
print(price_list)
price_list = list(map(int, price_list))
print(price_list)

print(type(price_list[0]))
```

```python
# Find the index of the row with the maximum value in column 'x'
max_row_index = np.argmax(price_list)

print(max_row_index)

# Extract the value from column 'y' in the row with the maximum value in column 'x'
y_value = name_list[max_row_index]

print(y_value)
print(price_list[max_row_index])
print(game_list[max_row_index])
```

```python
# Find the index of the row with the maximum value in column 'x'
min_row_index = np.argmin(price_list)

print(min_row_index)

# Extract the value from column 'y' in the row with the maximum value in column 'x'
y_value_min = name_list[min_row_index]

print(y_value_min)
print(price_list[min_row_index])
print(game_list[min_row_index])
```

```python
# Plot the histogram
plt.hist(price_list, bins=20, color='skyblue', edgecolor='black')  # Adjust the number of bins as needed
plt.xlabel('Price (in cents)')
plt.ylabel('Frequency')
plt.title('Distribution of the price')
plt.grid(True)
plt.xticks(np.arange(0, 6001, step=1000))
plt.savefig("Histogram_price.jpg")
plt.show()
```

```python
#plot the 9 images
plt.figure(figsize=(10, 6)) #size of the window

image_paths = [y_value_min,y_value]
print(image_paths)
i = 0
for path in image_paths:
    plt.subplot(1, 2, i + 1)  # Adjust subplot to 3x3 grid
    img = image.load_img("G:\\Dataset\\images\\"+path, target_size=(1080,1920))
    plt.imshow(img)
    plt.axis('off')
    i += 1

plt.tight_layout()  # Adjust spacing between subplots
plt.savefig("Images of games with the smallest (left) and largest (right) price")
plt.show()
```
