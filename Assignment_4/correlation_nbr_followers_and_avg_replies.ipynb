```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
```

```python
file_path = 'nbr followers and avg replies per politician.csv'

# Load the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
print(df.head())
```

```python
#compute Pearson correlation
correlation = df['nbrFollowers'].corr(df['averageReplies'])
print(correlation)
```

```python
plt.scatter(df['nbrFollowers'], df['averageReplies'], color = "blue")
#plt.xscale('log')
#plt.yscale('log')
plt.xlabel('Number of followers')
plt.ylabel('Average replies')
plt.title('Replies vs number of followers for every politician')
plt.show()
```

```python
# Scatter plot
plt.scatter(df['nbrFollowers'], df['averageReplies'], color="blue")

# Draw red rectangle
rect = Rectangle((0, 0), 20000, 50, linewidth=1, edgecolor='red', facecolor='none')
plt.gca().add_patch(rect)

# Labels and title
plt.xlabel('Number of followers')
plt.ylabel('Average replies')
plt.title('Replies vs number of followers for every politician')

# Show plot
plt.show()
```

```python
plt.scatter(df['nbrFollowers'], df['averageReplies'], color = "blue")
#plt.xscale('log')
#plt.yscale('log')
plt.xlabel('Number of followers')
plt.ylabel('Average replies')
plt.title('Replies vs number of followers for every politician (zoomed in)')
plt.xlim(0, 20000)
plt.ylim(0, 50)
plt.show()
```
